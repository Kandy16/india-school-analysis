```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('new-data-set/faculty-strength-quality-measure/#teachers-school.csv')
data.head()
```

```python
data[data['State/UT']=='Andaman and Nicobar Islands']
```

```python
data = data[['State/UT','Total No. of Teachers in all Schools - Total - 2008-09']]
data.head()
```

```python
data.rename(columns={'Total No. of Teachers in all Schools - Total - 2008-09':'#teachers-100-people'},index=str,inplace=True)
data.head()
```

```python
data[data['#teachers-100-people'].isnull()]
```

```python
demography = pd.read_csv('india-age-demography/state-wise-demography-students.csv')
demography.head()
```

```python
for state in list(demography['State/UT']):
    print(state,end=None)

    #print(demography['#Expected Students'][demography['State/UT']==state])

    temp = data[data['State/UT']==state]['#teachers-100-people']
    exp_temp = demography[demography['State/UT']==state]['#Expected Students']
    #print(temp.iloc[0])
    #print(exp_temp.iloc[0])
    #data[data['State/UT']==state].loc[0,'#teachers'] = temp.iloc[0] *100 / exp_temp.iloc[0]
    #print(temp.iloc[0] *100 / exp_temp.iloc[0])
    #print(temp.iloc)
    data.loc[data['State/UT']==state,'#teachers-100-people'] = temp.iloc[0] *100 / exp_temp.iloc[0]
```

```python
data.sort_values('State/UT',inplace=True)
data.head()
```
