```python
import pandas as pd
import numpy as np
import matplotlib
%matplotlib inline
```

```python
data_list = []

schools = pd.read_csv('#schools.csv')
data_list.append(schools)

teachers = pd.read_csv('#teachers.csv')
data_list.append(teachers)

students = pd.read_csv('#students.csv')
data_list.append(students)

students_pass = pd.read_csv('students-pass-percentage.csv')
data_list.append(students_pass)

students_dropout = pd.read_csv('school-drop-out-rate.csv')
data_list.append(students_dropout)

literacy_rate = pd.read_csv('literacy-rate.csv')
data_list.append(literacy_rate)

trained_teachers = pd.read_csv('trained-teachers-percentage.csv')
data_list.append(trained_teachers)

fulltime_teachers = pd.read_csv('full-time-teachers-percentage.csv')
data_list.append(fulltime_teachers)

govt_schools = pd.read_csv('govt-schools-percentage.csv')
data_list.append(govt_schools)

private_schools = pd.read_csv('private-schools-percentage.csv')
data_list.append(private_schools)

drinking_water = pd.read_csv('drinking-water-schools-percentage.csv')
data_list.append(drinking_water)

hand_wash = pd.read_csv('hand-wash-facility-percentage.csv')
data_list.append(hand_wash)

toilet_facility = pd.read_csv('toilet-facility-schools-percentage.csv')
data_list.append(toilet_facility)

electricity_facility = pd.read_csv('electricitiy-connection-schools-percentage.csv')
data_list.append(electricity_facility)

boundary_wall = pd.read_csv('boundary-wall-schools-percentage.csv')
data_list.append(boundary_wall)

playground_schools = pd.read_csv('play-ground-school-percentage.csv')
data_list.append(playground_schools)

pukka_schools = pd.read_csv('pucca-building-type-schools-percentage.csv')
data_list.append(pukka_schools)

computer_schools = pd.read_csv('computer-availability-schools-percentage.csv')
data_list.append(computer_schools)

midday_meal = pd.read_csv('mid-day-meal-school-percentage.csv')
data_list.append(midday_meal)

text_books = pd.read_csv('text-books-received-schools-percentage.csv')
data_list.append(text_books)
```

```python
col_list = []
col_list.append('State/UT')

temp = []
for data in data_list:
    temp_list = list(data.columns)
    for tmp in temp_list:
        temp.append(tmp)
temp = list(set(temp))
temp.remove('State/UT')

for tmp in temp:
    col_list.append(tmp)
    
col_list
```

```python
final_data = pd.DataFrame(columns=col_list)
state_list = list(data_list[0]['State/UT'])
state_list
```

```python
col_list.remove('State/UT')
for state in state_list:
    content = []
    content.append(state)
    for attr in col_list:
        for data in data_list:
            temp_list = list(data.columns)
            if(attr in temp_list):
                value = data[data['State/UT']==state][attr].iloc[0]
                content.append(value)
    final_data.loc[final_data.shape[0]] = content
final_data.head()
    
```

```python
final_data.sort_values('State/UT',inplace=True)
final_data.head()
```

```python
final_data.to_csv('final-data.csv',index=False)
```
